# General example of image opening and handling

```python
import image_viewer_class
import os
from astropy.coordinates import Angle
%matplotlib qt
```

```python
object = 'ZTF25acegldq'

iv = image_viewer_class.image_viewer('test_images', folder_list = os.listdir('test_images'))
iv.df_files[iv.df_files['object'] == object]
# coords = (Angle())

iv.view_image([4,9],
              manipulation_kw = {
                  'centered' : ('14:06:20.746h', '+39:43:14.51d'),
                  'zoom' : '0 3 0 d',
                  'stretch' : 'linear',
                #   'percentile' : (None, 99)
                #   'vminmax' : (0, 300)
              },
              plotting_kw={
                    'add_circle' : {
                            'center': ('14:06:20.746h', '+39:43:14.51d'),
                            'size' : '0 0 10 d'}
                })
```
