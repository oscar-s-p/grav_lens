```python
# imports
import numpy as np
import pandas as pd
import astropy.io.fits as fits
import matplotlib.pyplot as plt
```

```python
def load_fits_data(file_path: str) -> np.ndarray:
    """Returns the image data array of a FITS file stored in a path.

    :param file_path: A string with the path of the image to be opened.
    :return: The image retrieved from the path, as a bidimensional array,
            and the header.
    """

    with fits.open(file_path) as hdul:
        data = hdul[0].data.astype(np.float32)
        heads = hdul[0].header
    hdul.close()
    return data, heads
```
