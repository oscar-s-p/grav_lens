```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1,11)
R = np.linspace(0,4,100)

def sersic(n, R):
    bn = 2*n - 1/3
    bn = 1
    return np.exp(-bn * (R**n-1))

fig, ax = plt.subplots()
for ni in n:
    plt.plot(R, sersic(ni, R), label = ni)

ax.set_xlabel('$R/R_e$')
ax.set_ylabel('$I/I_e (R/R_e)$')
ax.legend()
ax.set_title('Srsic profiles with different $n$ indexes')
```
